```python
plt.figure(figsize=(16, 6))
sns.lineplot(data = flight_data)
```

```python
plt.figure(figsize= (10,5))

plt.title('Average arrival delay for MQ airlines flights by months')

sns.barplot(x= flight_data.index, y=flight_data.MQ)

plt.ylabel('Arrival delay(in minutes)')
```

```python
plt.figure(figsize=(14,7))

plt.title('Average arrival delay for all airlines flights by months')

sns.heatmap(data= flight_data, annot = True)

plt.xlabel('Airlines')
```
